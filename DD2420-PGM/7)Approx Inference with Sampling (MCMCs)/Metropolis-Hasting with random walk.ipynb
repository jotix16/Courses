```python
import numpy as np
import math
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
def metropolis_hasting(target,sigma_prop=1, iter = 10000):
    '''
    metropolis hasting algorithm with gaussian distributed random walk
    '''
    samples = np.zeros(iter)
    x = 0 # starting state
    acceptance = 0
    np.random.seed(100)
    for i in range(iter):
        x_star = x + np.random.normal(scale = sigma_prop, size=1)
        if np.random.rand() < p(x_star) / p(x):
            x = x_star
            acceptance += 1
        samples[i] = x
    return samples,acceptance/iter
```

```python
def p(x):
    '''
    target distribution up to a constant factor
    '''
    a1 = a2 = 0.5
    mu1 = 0
    sigma1 = 1
    mu2 = 3
    sigma2 = 0.4
    
    return a1*st.multivariate_normal.pdf(x,mu1,sigma1) + a2*st.multivariate_normal.pdf(x,mu2,sigma2)
```

```python
def p_true(x):
    '''
    the true target distribution
    '''
    a1 = a2 = 0.5
    mu1 = 0
    sigma1 = 1
    mu2 = 3
    sigma2 = 0.4
    
    return (a1*st.multivariate_normal.pdf(x,mu1,sigma1) + a2*st.multivariate_normal.pdf(x,mu2,sigma2))/(a1+a2)
```

```python
sigma_proposals = [0.1,0.3,0.7,1,1.5,2,7,20]
iter = 10000

# Plot the distribution curve
for i in range(len(sigma_proposals)):
    samples, acceptance_ratio = metropolis_hasting(p,sigma_proposals[i],iter)
    #plt.subplot(3,2,i+1)
    plt.title("Variance: "+str(sigma_proposals[i])+". Acceptance ratio: "+str(acceptance_ratio))
    count, bins, ignored = plt.hist(samples,20,density=True)
    plt.plot(bins,p_true(bins),color = 'r',label="Target distribution")
    plt.xlabel("x")
    plt.legend(loc='upper right', frameon=False)
    plt.show()
```
